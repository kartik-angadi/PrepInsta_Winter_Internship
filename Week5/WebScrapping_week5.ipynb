#Web Scrapping of the website
#Books to Scrape We love being scraped!


```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create an empty list to store book details
books_details = []

# Loop through pages 1 to 4
for i in range(1, 5):
    # Construct the URL for each page
    url = f"http://books.toscrape.com/catalogue/page-{i}.html"

    # Send an HTTP request to the URL and get the content
    response = requests.get(url)
    response = response.content

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response, 'html.parser')

    # Check if 'ol' element with class 'row' exists
    ol = soup.find('ol', class_='row')

    if ol:
        # If 'ol' element is found, find all 'article' elements with class 'product_pod'
        articles = ol.find_all('article', class_='product_pod')

        # Loop through each book in the page
        for article in articles:
            # Find the 'img' element to get the book title
            image = article.find('img')
            title = image.attrs['alt']

            # Find the 'p' element to get the star rating (class is the second element of the class list)
            star = article.find('p')['class'][1]

            # Find the 'p' element with class 'price_color' to get the book price
            price = article.find('p', class_='price_color').text

            # Append the book details to the list
            books_details.append([title, star, price])
    else:
        # Print a message if 'ol' element is not found on the page
        print(f"No 'ol' element with class 'row' found on page {i}")

# Print the scraped details
for details in books_details:
    print(details)

# Create a DataFrame using pandas to store the scraped details
data_frame = pd.DataFrame(books_details, columns=['Title', 'Rating', 'Price'])
```

```python
#creating csv file to store the details
data_frame.to_csv('books_detailes.csv')
data_frame
```
